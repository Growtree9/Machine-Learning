>step1:Gathering data

```python
#importing packages
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/insurance.csv')
```

```python
data.head()
```

```python
data.info()
```

>step2:Data preprocessing

```python
#one hot encoding-convert the cogegorical variables into numerical variables
data['sex']=data['sex'].apply({'male':1,'female':0}.get)
```

```python
data.head()
```

```python
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
data['region']=data['region'].apply({'southeast':1,'southwest':0,'northeast':2,'northwest':3}.get)
```

```python
data.head()
```

>step3:Dividing data into dependent and independent variables

```python
x=data[['age','sex','bmi','children','smoker','region']]#independent variables
y=data[['expenses']]#dependent variables
```

>step4:spliting data into training and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
#test_size=0.2 means 20% of data will be used for testing the accuracy of model
#80% of data be used for making the model learn
```

>step5:creating ML model using Linear Regression algorithm

```python
#import Linear Regression algorithm
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()#activating the algorithm for use
```

```python
#training the model using .fit()
linear_model.fit(x_train,y_train)
```

```python
y_pred=linear_model.predict(x_test)#predict()used for predicting future values
```

```python
print(y_pred)#y_pred is amount of health insurance as calculated by ML model
```

```python
#checking accuracy of the model
accuracy=linear_model.score(x,y)
```

```python
print('the accuracy of ML model using Linear Regression algorithm is',accuracy)
```
